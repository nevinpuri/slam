```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
from fastai.data.transforms import get_image_files
from fastai.vision.data import SegmentationDataLoaders
from fastai.vision.all import unet_learner
from fastai.vision.models import resnet34
```

```python
import fastai
print(fastai.__version__)
```

```python
train_images = get_image_files("/media/nevin/Trash Games1/selfdrive/train")
```

```python
def label_func(img):
    return f"/media/nevin/Trash Games1/selfdrive/train_label/{img.stem}_label{img.suffix}"

train_label = np.array(cv.imread(str(label_func(train_images[0]))))
print(np.unique(train_label, return_counts=True))
```

```python
print(train_images[0].stem)
```

```python
img = cv.imread(str(train_images[0]))
plt.imshow(img)
```

```python
codes = ["nothing", "left", "right"]
```

```python
dls = SegmentationDataLoaders.from_label_func(
    path="/media/nevin/Trash Games1/selfdrive/",
    bs=8,
    fnames=train_images,
    label_func=label_func,
    codes=codes
)
```

```python
dls.show_batch(max_n=6)
```

```python
model = unet_learner(dls, resnet34)
model.fine_tune(8)
```
