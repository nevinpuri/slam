```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
from fastai.data.transforms import get_image_files
from fastai.vision.data import SegmentationDataLoaders
from fastai.vision.all import unet_learner, Learner
from fastai.vision.models import resnet18
from fastai.metrics import DiceMulti
from fastseg import MobileV3Small
```

```python
import fastai
print(fastai.__version__)
```

```python
train_images = get_image_files("/media/nevin/Trash Games1/selfdrive/train")
```

```python
def label_func(img):
    return f"/media/nevin/Trash Games1/selfdrive/train_label/{img.stem}_label{img.suffix}"

train_label = np.array(cv.imread(str(label_func(train_images[0]))))
print(np.unique(train_label, return_counts=True))
print(label_func(train_images[0]))
```

```python
print(train_images[0])
```

```python
img = cv.imread(str(train_images[0]))
width = int(img.shape[1] * 0.2)
height = int(img.shape[0] * 0.2)

dim = (width, height)
resized = cv.resize(img, dim, interpolation=cv.INTER_AREA)
plt.imshow(resized)
```

```python
codes = ["nothing", "left", "right"]
```

```python
dls = SegmentationDataLoaders.from_label_func(
    path="/media/nevin/Trash Games1/selfdrive/",
    bs=4,
    fnames=train_images,
    label_func=label_func,
    codes=codes
)
```

```python
dls.show_batch(max_n=6)
```

```python
model = MobileV3Small(num_classes=3, use_aspp=True)
learn = Learner(dls, model, metrics=[DiceMulti()])
```

```python
learn.fine_tune(8)
```

```python
learn.export("/media/nevin/Trash Games1/selfdrive/model.pkl")
```

```python
print(dir(learn))
```

```python
img = cv.imread(str(train_images[0]))
plt.imshow(learn.predict(img)[0])
```
